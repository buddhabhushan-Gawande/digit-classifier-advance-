```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np

digit_clasifier = tf.keras.models.load_model("number classifer")
```

```python
# load img
cap = cv.VideoCapture(0)
numbers = ["Zero","One","Two","Three","Four","Six","Six","Seven","Eigit","Nine"]
while True:
    truefalse,frame = cap.read()
    if truefalse == False:
        break
    frame = cv.resize(frame,(400,300))

    
    
    # convert to grayscale
    gray = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    # edge detectiion insted of throsholding
    edge = cv.Canny(frame,80,255)
    # find contours
    contours,hierarchy = cv.findContours(edge,cv.RETR_EXTERNAL,cv.CHAIN_APPROX_SIMPLE)

    for cont in contours:
        # print(cv.contourArea(cont))

        if cv.contourArea(cont) < 100 and cv.contourArea(cont) > 10:
            print(cv.contourArea(cont))
            # cv.drawContours(frame,cont,-1,(0,0,255),2)
            # cv.imshow("final",frame)
            x1,y1,l,h = cv.boundingRect(cont)
            # center point of digit
            (x2,y2),radius = cv.minEnclosingCircle(cont)
            x2 = int(x2)
            y2 = int(y2)
            radius = int(radius*1.5)
            # points for rectrangle digit corp
            p1 = x2-radius
            p2 = y2-radius
            p3 = x2+radius
            p4 = y2+radius

            # extracting each digit from image
            digit = edge[p2:p4,p1:p3]
            if digit.shape >= (20,20) and (digit.shape[0] == digit.shape[1]):

                # preparing for predict
                digit = cv.resize(digit,(28,28))
                x,digit = cv.threshold(digit,50,255,0)
                # cv.imshow("samp",digit)
                digit = digit.reshape(-1)
                digit = tf.keras.utils.normalize(digit,axis=0)
                pred = digit_clasifier.predict(digit)
                max_value = np.max(pred)
                locate_max_value = np.where(pred == max_value)
                prdted_dgt = locate_max_value[1]
                
                # showing reshults
                pf2 = int(p2-10)
                cv.putText(frame,numbers[int(prdted_dgt)],(p1,pf2),cv.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),2)
                cv.rectangle(frame,(p1,p2),(p3,p4),(0,255,0),2)
                cv.imshow("final",frame)
    if cv.waitKey(1) == ord('q'):
        break
cv.waitKey(0)
cv.destroyAllWindows()
cap.release()
```
